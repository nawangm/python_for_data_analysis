```python
data=np.random.rand(20)  #uniformly distributed
pd.cut(data,4,precision=2)  # cut into 4 bins equal-length bins,precise to 2 digits before 0
```

A closely related function, qcut, bins the data based on sample quantiles.Depending
on the distribution of the data, using cut will not usually result in each bin having the
same number of data points. Since **qcut** uses sample quantiles instead, by definition
you will obtain roughly equal-size bins:

```python
data = np.random.randn(1000)   #normally distributed
cats=pd.cut(data,4)
cats
```

```python
pd.value_counts(cats)
```

```python
#Similar to cut you can pass your own quantiles (numbers between 0 and 1, inclusive):
pd.qcut(data,[0,0.1,0.5,0.9,1])
```

```python
import re
```

```python
text = "foo bar\t baz \tqux"
re.split('\s+',text)
```

```python
#or  re first needs to compile, then split
regex=re.compile('\s+')
regex.split(text)
```

```python
#you wanted to get a list of all patterns matching the regex, you can use the findall method:
regex.findall(text)
```

match and search are closely related to findall. While findall returns all matches
in a string, search returns only the first match. More rigidly, match only matches at
the beginning of the string

```python
text = """Dave dave@google.com
Steve steve@gmail.com
Rob rob@gmail.com
Ryan ryan@yahoo.com
"""
pattern=r'[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,4}'
# re.IGNORECASE makes the regex case-insensitive
regex=re.compile(pattern,flags=re.IGNORECASE)
```

```python
regex.findall(text)
```

```python
m=regex.search(text)
m
```

```python
text[m.start():m.end()]
```

regex.match returns None, as it only will match if the pattern occurs at the start of the
string:

```python
print(regex.match(text))
```

sub will return a new string with occurrences of the pattern replaced by the
a new string:

```python
regex.sub('REDACTED',text)
```

Suppose you wanted to find email addresses and simultaneously segment each
address into its three components: username, domain name, and domain suffix. To
do this, put parentheses around the parts of the pattern to segment:

```python
pattern=r'([A-Z0-9._%+-]+)@([A-Z0-9.-]+)\.([A-Z]{2,4})'
regex=re.compile(pattern,flags=re.IGNORECASE)
m=regex.match('wesm@bright.net')
m
```

```python
m.groups()
```

```python
#findall returns a list of tuples when the pattern has groups:
regex.findall(text)
```

```python
#sub \1 first matched group, \2 2nd
print(regex.sub(r'Username: \1,Domain: \2, Suffix: \3',text))
```

concat function arguments

objs List or dict of pandas objects to be concatenated; this is the only required argument  
axis Axis to concatenate along; defaults to 0 (along rows)  
join Either 'inner' or 'outer' ('outer' by default); whether to intersection (inner) or union
(outer) together indexes along the other axes  
join_axes Speciic indexes to use for the other n1 axes instead of performing union/intersection logic  
keys Values to associate with objects being concatenated, forming a hierarchical index along the
concatenation axis; can either be a list or array of arbitrary values, an array of tuples, or a list of
arrays (if multiple-level arrays passed in levels)  
levels Speciic indexes to use as hierarchical index level or levels if keys passed  
names Names for created hierarchical levels if keys and/or levels passed  
verify_integrity Check new axis in concatenated object for duplicates and raise exception if so; by default (False)
allows duplicates  
ignore_index Do not preserve indexes along concatenation axis, instead producing a new range(total_length) index

Example: Group-Wise Linear Regression

use groupby to perform more
complex group-wise statistical analysis, as long as the function returns a pandas
object or scalar value. e.g define the following regress function
(using the statsmodels econometrics library), which executes an ordinary least
squares (OLS) regression on each chunk of data:

```python
import statsmodels.api as sm
import statsmodels.api as sm
def regress(data, yvar, xvars):
    Y = data[yvar]
    X = data[xvars]
    X['intercept'] = 1.
    result = sm.OLS(Y, X).fit()
    return result.params
```

```python
rets.groupby(get_year).apply(regress,'AAPL',['SPX'])
```

```python
Since the dataset is split into files by year, one of the first things to do is to assemble
all of the data into a single DataFrame and further to add a year field
```

```python
years=range(1880,2011)
pieces=[]
for year in years:
    df=pd.read_csv('pydabook_2ed/datasets/babynames/yob{}.txt'.format(year),sep=',',header=None,names=['name', 'sex', 'births'])
    df['year']=year
    pieces.append(df)
pieces[0].head()
```

```python
# Concatenate everything into a single DataFrame
names=pd.concat(pieces,ignore_index=True)
names[:20]
```

```python
#aggregating the data at the year and sex level using groupby or pivot_table
names.groupby(['year','sex'])['births'].sum()
```

```python
total_births=pd.pivot_table(names,index='year',columns='sex',values='births',aggfunc='sum')
```

```python
total_births.plot()
```

```python
def add_prop(group):
    group['prop']=group.births/group.births.sum()
    return group
names=names.groupby(['year','sex']).apply(add_prop)
names.head()
```

```python
names.groupby(['year','sex']).prop.sum()
```

```python
#extract the top 1,000 names for each sex/year combination
def get_top1000(group):
    return group.sort_values(by='births',ascending=False)[:1000]

grouped=names.groupby(['year','sex']).apply(get_top1000)
grouped.head()
# Drop the group index, not needed
grouped.reset_index(inplace=True,drop=True)
grouped.head()
```

```python
# or
pieces=[]
for year, group in names.groupby(['year','sex']):
    grouped=group.sort_values(by='births',ascending=False)[:1000]
    pieces.append(grouped)
        
top1000=pd.concat(pieces,ignore_index=True)   
top1000.head()
```
