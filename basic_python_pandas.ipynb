**for loop**  
for loops are for iterating over a collection (like a list or tuple) or an iterater. The

for value in collection:

```python
#for loop
#continue, # break
sequence = [1, 2, None, 4, None, 5]
total = 0
for value in sequence:
    if value is None:
        continue
    total += value
print(total)
```

```python
words = ['apple', 'bat', 'bar', 'atom', 'book','124']
by_letter={}
for word in words:
    letter=word[0]
    if letter not in by_letter:
        by_letter[letter]=[word]
    else:
        by_letter[letter].append(word)
print(by_letter)
```

```python
#The setdefault dict method is for precisely this purpose. The preceding for loop can be rewritten as:
for word in words:
    letter = word[0]
    by_letter.setdefault(letter, []).append(word)
#The built-in collections module has a useful class, defaultdict, which makes this even easier. To create one, 
#you pass a type or function for generating the default value for each slot in the dict:
from collections import defaultdict
by_letter = defaultdict(list)
for word in words:
    by_letter[word[0]].append(word)
```

An iterator is any object that will yield objects to the Python interpreter when used in a context like a for loop.   
**A generator** is a concise way to construct a new iterable object.  
To create a generator, use the **yield** keyword instead of return in a function:

```python
def square(n=10):
    print('generating squares from 1 to {0}'.format(n**2))
    for i in range(1, n+1):
        yield i**2
gen = square()
gen
```

```python
for x in gen:
    print(x, end=' ')
```

Generator expresssions  
a generator analogue to list, dict, and set comprehensions; to create one,

```python
gen = (x**2 for x in range(10))
gen
```

```python
#same with:
def _make_gen():
    for x in range(10):
        yield x**2
    
gen =_make_gen()
gen
```

```python
import itertools
first_letter = lambda x:x[0]
names = ['Alan', 'Adam', 'Wes', 'Will', 'Albert', 'Steven']
for letter, names in itertools.groupby(names, first_letter):
    print(letter, list(names))
    
```

**Random Walk**

```python
import random
```

```python
walk=[]
position=0
steps=1000
for step in range(steps):
    step=1 if random.randint(0,1) else -1
    position+=step
    walk.append(position)
plt.plot(walk[:100])
plt.show()
```

help(np.random)
Compatibility functions
    ==============================================================================
    rand                 Uniformly distributed values.
    randn                Normally distributed values.

```python
# or
nsteps=1000
draws=np.random.randint(0,2,size=nsteps)
steps=np.where(draws>0,1,-1)
walk=steps.cumsum()
walk.min()
walk.max()
(np.abs(walk)>=10).argmax()
```

```python
#simulating many random walks --5000 walks
nwalks=5000
nsteps=1000
draws=np.random.randint(0,2,size=(nsteps,nwalks))
steps=np.where(draws>0,1,-1)
walks=steps.cumsum(1)
plt.plot(walks[:,100])
plt.show()
```

```python
crossing_times = (np.abs(walks[hits30]) >= 30).argmax(1)  # first index reaching (argmax)
crossing_times.mean()
```

CSV files come in many different flavors. To define a new format with a different
delimiter, string quoting convention, or line terminator, we define a simple subclass
of **csv.Dialect**:

```python
class my_dialect(csv.Dialect):
    lineterminator='\n'
    delimiter=';'
    quotechar='"'
    quoting =csv.QUOTE_MINIMAL
    
#reader=csv.reader(f, dialect=my_dialect)
#individual opetion
#reader=csv.reader(f, delimiter='|')
```

CSV dialect options  
delimiter One-character string to separate ields; defaults to ','.  
lineterminator Line terminator for writing; defaults to '\r\n'. Reader ignores this and recognizes cross-platform
line terminators.  
quotechar Quote character for ields with special characters (like a delimiter); default is '"'  
quoting Quoting convention. Options include csv.QUOTE_ALL (quote all ields), csv.QUOTE_MINIMAL (only ields with special characters like the delimiter), csv.QUOTE_NONNUMERIC, and csv.QUOTE_NONE (no quoting). See Python’s documentation for full details. Defaults to QUOTE_MINIMAL.
skipinitialspace  Ignore whitespace after each delimiter; default is False.  
doublequote How to handle quoting character inside a ield; if True, it is doubled (see online documentation
for full detail and behavior).  
escapechar String to escape the delimiter if quoting is set to csv.QUOTE_NONE; disabled by default

**csv.writer**

```python
with open('mydata.csv','w') as f:
    writer=csv.writer(f, dialect=my_dialect)
    writer.writerow(('1', '2', '3'))
    writer.writerow(('4', '5', '6'))
    writer.writerow(('7', '8', '9'))
```

**HDF5**

The “HDF” in HDF5 stands for hierarchical data format. Each HDF5 file can store multiple datasets and support‐
ing metadata. Compared with simpler formats, HDF5 supports on-the-fly compres‐
sion with a variety of compression modes, enabling data with repeated patterns to be
stored more efficiently. HDF5 can be a good choice for working with very large data‐
sets that don’t fit into memory, as you can efficiently read and write small sections of
much larger arrays.


```python

frame = pd.DataFrame({'a': np.random.randn(100)})
type(frame)
store=pd.HDFStore('mydata.h5')
store['obj1']=frame
store['obj1_col']=frame['a']
store
```

```python
#store['obj1']
```

```python
store.put('obj2',frame, format='table')
#store['obj2']
```

```python
store.select('obj2',where=['index >=10 and index <=15'])
```

```python
store.close()
```

```python
frame.to_hdf('mydata.h5','obj3',format='table')
```

```python
pd.read_hdf('mydata.h5','obj3', where=['index <5'])
```

HDF5 is not a database. It is best suited for write-once, read-many
datasets. While data can be added to a file at any time, if multiple
writers do so simultaneously, the file can become corrupted.

```python
query="""CREATE TABLE test(a VARCHAR(20), b VARCHAR(20), c REAL, d INTEGER);"""
con=sqlite3.connect('mydata.sqlite')
con.execute(query)
```

```python
data=[('Altanta','Georgia',1.25,6),('Tallahassee', 'Florida', 2.6, 3),('Sacramento', 'California', 1.7, 5)]
```

```python
stmt='INSERT INTO test VALUES(?,?,?,?)'
con.executemany(stmt,data)
```

```python
con.commit()
```

Most Python SQL drivers (PyODBC, psycopg2, MySQLdb, pymssql, etc.) return a list
of tuples when selecting data from a table:

```python
cursor=con.execute('select * from test')
rows=cursor.fetchall()
rows
```

```python
cursor.description
```

You can pass the list of tuples to the DataFrame constructor, but you also need the
column names, contained in the cursor’s description attribute:

```python
pd.DataFrame(rows, columns=[x[0] for x in cursor.description])
```

The SQLAlchemy project is a popular Python SQL toolkit that abstracts away many of the common differences
between SQL databases.

```python
import sqlite3
query="""CREATE TABLE test4(a VARCHAR(20), b VARCHAR(20), c REAL, d INTEGER);"""
con=sqlite3.connect('mydat.sqlite')
con.execute(query)
data=[('Altanta','Georgia',1.25,6),('Tallahassee', 'Florida', 2.6, 3),('Sacramento', 'California', 1.7, 5)]
stmt='INSERT INTO test4 VALUES(?,?,?,?)'
con.executemany(stmt,data)
con.commit()
```

```python
import sqlalchemy as sqla
db=sqla.create_engine('sqlite:///mydat.sqlite')
pd.read_sql('select * from test4',db)
```
